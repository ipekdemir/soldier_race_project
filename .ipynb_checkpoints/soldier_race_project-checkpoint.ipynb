# Determines

The 2012 US Army Anthropometric Survey (ANSUR II) was executed by the Natick Soldier Research, Development and Engineering Center (NSRDEC) from October 2010 to April 2012 and is comprised of personnel representing the total US Army force to include the US Army Active Duty, Reserves, and National Guard. In addition to the anthropometric and demographic data described below, the ANSUR II database also consists of 3D whole body, foot, and head scans of Soldier participants. These 3D data are not publicly available out of respect for the privacy of ANSUR II participants. The data from this survey are used for a wide range of equipment design, sizing, and tariffing applications within the military and has many potential commercial, industrial, and academic applications.

The ANSUR II working databases contain 93 anthropometric measurements which were directly measured, and 15 demographic/administrative variables explained below. The ANSUR II Male working database contains a total sample of 4,082 subjects. The ANSUR II Female working database contains a total sample of 1,986 subjects.

# Modelling
- Fit the model with train dataset
- Get predict from vanilla model on both train and test sets to examine if there is over/underfitting   
- Apply GridseachCV for both hyperparemeter tuning and sanity test of our model.
- Use hyperparameters that you find from gridsearch and make final prediction and evaluate the result according to chosen metric.

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

## 1. Logistic model

### Vanilla Logistic Model

```python
cat = X_train.select_dtypes("object").columns
cat 
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler

column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",
                                                      sparse=False), cat),
                                                      remainder=MinMaxScaler(),
                                                      verbose_feature_names_out=False)
```

```python
from sklearn.pipeline import Pipeline
operations = [("OneHotEncoder", column_trans),#dönüşüm yapacak
              ("log", LogisticRegression(class_weight='balanced', #modeli uygulayacak, class weight=balanced ı gridsearch param grid de kullnama maliyet artar burda kullan
                                         random_state=101))] #random state ı burda göster çünkü log solverın arka planda sag ve saga seçersek skorlar değişiyor o yüzden belirtmek zorundayız

pipe_log_model = Pipeline(steps=operations)
```

```python
pipe_log_model.fit(X_train,y_train)
eval_metric(pipe_log_model, X_train, y_train, X_test, y_test)
```

```python
scoring = {"precision_Hispanic" : make_scorer(precision_score, average = None, labels = ["Hispanic"]),   #hedef label hispanic eğer tek class olmasaydı average ın karşsına farklı şeyler yazardık.none yerine weighted micro macro kullanabilirdik micro görünmüyor accuracy ile aynı.accuracy tüm gözlemlere aynı davranıyor weighted çoğunluğu daha fazala ağırlıklandırıyor.macro skoru düşük olana yakınsıyor
           "recall_Hispanic" : make_scorer(recall_score, average = None, labels =["Hispanic"]),
           "f1_Hispanic" : make_scorer(f1_score, average = None, labels =["Hispanic"])}
```

```python
operations = [("OneHotEncoder", column_trans), 
              ("log", LogisticRegression(class_weight='balanced',
                                         random_state=101))]

model = Pipeline(steps=operations)


scores = cross_validate(model,
                        X_train, 
                        y_train, 
                        scoring=scoring,
                        cv=10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]   #skorlar birbirine yakın ama varyansı daha da düşürmeye çalışıyoruz genellemeyi artırmak için
```

### Logistic Model GridsearchCV

```python
recall_Hispanic =  make_scorer(recall_score, average=None, labels=["Hispanic"]) #gridsearche recallü artırsın diye işlem yapacağız
#recall seçmemizin nedeni müşter isteğine göre model yüzde yüz bilsin diyorsak amacımız precision ı max ayırmaya çalışırız.orantılamalıyız
```

```python
param_grid = {
    "log__C": [0.1, 0.2 , 0.5],  #c de küçük değerleri de aldık varyansı azaltmaya çalıştığımız için aslında parameterelerimiz çok olduğundan c nin içine iki üç değer versek iyi olur maliyet açısından
    'log__penalty': ["l1", "l2"],
    'log__solver': ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga'], #penalty ve solver eşleşmesinde bazıları birbiryler kullanılmıyor gridsearch bunları dikkate alıyor
}
#grid searche verilen değerler overfite götürmemeli çünkü gridsearch overfit kontrolü yapmaz.default değerlere bak cv de kontrol yap
```

```python
operations = [("OneHotEncoder", column_trans), 
              ("log", LogisticRegression(class_weight='balanced',
                                         random_state=101))]

model = Pipeline(steps=operations)


log_model_grid = GridSearchCV(model,
                              param_grid,
                              scoring=recall_Hispanic, #hispaniclerin recallünü max yapacak şekilde hyper parameterları bul
                              cv=10,
                              n_jobs=-1,
                              return_train_score=True)
```

```python
log_model_grid.fit(X_train,y_train) #modelimizi eğittik
```

```python
log_model_grid.best_estimator_ #best estimatorlarına baktık
```

```python
pd.DataFrame(log_model_grid.cv_results_).loc[log_model_grid.best_index_, ["mean_test_score", "mean_train_score"]] #burada varyans azalmış mı ona bakıyoruz ama classification özelinde sadece varyans ile overfite karar veremeyiz sadece recall sadece f1 e bakarak karar veremeyiz hem precision hem recalle bakmalıyız gridsearch sonrası alacağımı değerler için tekrar bakmalıyız 
```

```python
eval_metric(log_model_grid, X_train, y_train, X_test, y_test)
```

```python
operations = [("OneHotEncoder", column_trans), 
              ("log", LogisticRegression(C=0.5,   #best hyperlarlara baktık kontrol amaçlı overfit var mı yok mu
                                         class_weight='balanced',
                                         random_state=101))]

model = Pipeline(steps=operations)


scores = cross_validate(model,
                        X_train, 
                        y_train, 
                        scoring=scoring,
                        cv=10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

##  Smote implement

resampling metodu datamızdaki gözlem sayısını değiştirerek datamızın yapısını değiştirme. ya artıracağız ya azaltacağız iki mesudu var over sampling azınlık olan sınıfın sayısını çoğunluğa yaklaştırıyor ya da aynı yapıyor diğeri de tam tersi
ama handikapları var oversampling yazarken overfit yaparız o yüzden azar azar artırma yapmalıyız undersampling de ise datadan veri kaybı yaşıyrouz o yüzden undersampling yaprken maks datanın %20-25 ine kadar yapmalıyız hala düzelmiyorsa underı kullanmay devam etme çünkü veri kaybı çok olacak dünyadan uzaklaşmış olacağız over under yaptık düzeltemedik bitiriyoruz illa yapmak zorudnda değiliz ayrıca KESİNLİKLE ÖNCE TRAİNE AYIRIP SDCNTRAİN TESTE UNDER OVER A YAPACAĞIZ ÖNCE AYIRACAğIZ TRAIN-TEST, SONRA TRAINI ALIP UNDER/OVER YAPACAĞIZ.hispaniclerin data içinde sayısı az diğerleri fazla. bu endenle yeteri kadar obsevation yok  feature başına düşen sayı yetersiz bunu artırmak için oversampling/undersampling işlemleri
oversampling de hispanicelrin sayısını artırıyoruz.

```python
from imblearn.over_sampling import SMOTE # azınlık olan classları çoğunluk classa eşitler veya yakınlaştırır.
from imblearn.under_sampling import RandomUnderSampler # çoğunluk olan classı azınlık olan classa eşitler veya yakınlaştırır
from imblearn.pipeline import Pipeline as imbpipeline#imblearnin pipeline da aynı oldupun as verdik
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore", 
                                                      sparse=False), cat), 
                                       remainder=MinMaxScaler())
```

```python
X_train_ohe = column_trans.fit_transform(X_train)

# Önce manuel olarak X_train ve X_test' one hot encoder dönüşümü uyguluyoruz. 
```

```python
over = SMOTE()
X_train_over, y_train_over = over.fit_resample(X_train_ohe, y_train) #arka planda KNN kullanıyor

# over_sampling(smote) ve under_sampling (RandomUnderSampler) sadece X_train ve y_train'e uygulanır. tüm dataya uygulanıp sonra
# data train ve test setine ayrılırsa data_leakege gereçekleşir ve scorlarınız çok iyi olur. Bu sebeple data önce train ve 
# test setine ayrılmalı ve sonra da sadece train datasına uygulanmalıdır.
```

## Logistic Regression Over/ Under Sampling

Şimdi yukarda yaptığımız tüm dönüşümleri ve modele verme işlemlerini otomotize hale getirelim.

What does the smote pipeline do?

smote_pipeline.fit(X_train, y_train) 

--> column_trans.fit_transform(X_train) #(Onehooencoder and minmaxscaler for numeric features)

--> over.fit_resample(X_train_transform, y_train) 
                                     
--> under.fit_resample(X_train_transform_over, y_train_over)

--> log_model.fit(X_train_transform_over_under, y_train_over_under)


for predict, over and under sumpling algortims do nothing for X_test.

smote_pipeline.predict(X_test)

--> column_trans.transform(X_test) #(Onehotoencoder and minmaxscaler for numeric features) 

--> log_model.predict(X_test_transform)

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",
                                                      sparse=False), cat), 
                                       remainder=MinMaxScaler())
```

```python
operations = [("OneHotEncoder",column_trans), 
              ('o', over),
              ('u', under), 
              ("log", LogisticRegression(C=0.2, 
                                         random_state=101))] #("scaler", MinMaxScaler())

# over veya under sampling dataya uygulandığında kesinlikle class_weight="balanced" kullanılmaz. çnkğ yukarda istediğimiz sayıya getirdik bunu kulalnırsak yaptığımız işlekmin anlamı kalmaz
```
