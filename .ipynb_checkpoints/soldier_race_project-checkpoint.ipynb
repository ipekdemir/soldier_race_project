# Determines

The 2012 US Army Anthropometric Survey (ANSUR II) was executed by the Natick Soldier Research, Development and Engineering Center (NSRDEC) from October 2010 to April 2012 and is comprised of personnel representing the total US Army force to include the US Army Active Duty, Reserves, and National Guard. In addition to the anthropometric and demographic data described below, the ANSUR II database also consists of 3D whole body, foot, and head scans of Soldier participants. These 3D data are not publicly available out of respect for the privacy of ANSUR II participants. The data from this survey are used for a wide range of equipment design, sizing, and tariffing applications within the military and has many potential commercial, industrial, and academic applications.

The ANSUR II working databases contain 93 anthropometric measurements which were directly measured, and 15 demographic/administrative variables explained below. The ANSUR II Male working database contains a total sample of 4,082 subjects. The ANSUR II Female working database contains a total sample of 1,986 subjects.

# Modelling
- Fit the model with train dataset
- Get predict from vanilla model on both train and test sets to examine if there is over/underfitting   
- Apply GridseachCV for both hyperparemeter tuning and sanity test of our model.
- Use hyperparameters that you find from gridsearch and make final prediction and evaluate the result according to chosen metric.

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

## 1. Logistic model

### Vanilla Logistic Model

```python
cat = X_train.select_dtypes("object").columns
cat 
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler

column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",
                                                      sparse=False), cat),
                                                      remainder=MinMaxScaler(),
                                                      verbose_feature_names_out=False)
```

```python
from sklearn.pipeline import Pipeline
operations = [("OneHotEncoder", column_trans),#dnm yapacak
              ("log", LogisticRegression(class_weight='balanced', #modeli uygulayacak, class weight=balanced  gridsearch param grid de kullnama maliyet artar burda kullan
                                         random_state=101))] #random state  burda gster nk log solvern arka planda sag ve saga seersek skorlar deiiyor o yzden belirtmek zorundayz

pipe_log_model = Pipeline(steps=operations)
```

```python
pipe_log_model.fit(X_train,y_train)
eval_metric(pipe_log_model, X_train, y_train, X_test, y_test)
```

```python
scoring = {"precision_Hispanic" : make_scorer(precision_score, average = None, labels = ["Hispanic"]),   #hedef label hispanic eer tek class olmasayd average n karsna farkl eyler yazardk.none yerine weighted micro macro kullanabilirdik micro grnmyor accuracy ile ayn.accuracy tm gzlemlere ayn davranyor weighted ounluu daha fazala arlklandryor.macro skoru dk olana yaknsyor
           "recall_Hispanic" : make_scorer(recall_score, average = None, labels =["Hispanic"]),
           "f1_Hispanic" : make_scorer(f1_score, average = None, labels =["Hispanic"])}
```

```python
operations = [("OneHotEncoder", column_trans), 
              ("log", LogisticRegression(class_weight='balanced',
                                         random_state=101))]

model = Pipeline(steps=operations)


scores = cross_validate(model,
                        X_train, 
                        y_train, 
                        scoring=scoring,
                        cv=10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]   #skorlar birbirine yakn ama varyans daha da drmeye alyoruz genellemeyi artrmak iin
```

### Logistic Model GridsearchCV

```python
recall_Hispanic =  make_scorer(recall_score, average=None, labels=["Hispanic"]) #gridsearche recall artrsn diye ilem yapacaz
#recall sememizin nedeni mter isteine gre model yzde yz bilsin diyorsak amacmz precision  max ayrmaya alrz.orantlamalyz
```

```python
param_grid = {
    "log__C": [0.1, 0.2 , 0.5],  #c de kk deerleri de aldk varyans azaltmaya altmz iin aslnda parameterelerimiz ok olduundan c nin iine iki  deer versek iyi olur maliyet asndan
    'log__penalty': ["l1", "l2"],
    'log__solver': ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga'], #penalty ve solver elemesinde bazlar birbiryler kullanlmyor gridsearch bunlar dikkate alyor
}
#grid searche verilen deerler overfite gtrmemeli nk gridsearch overfit kontrol yapmaz.default deerlere bak cv de kontrol yap
```

```python
operations = [("OneHotEncoder", column_trans), 
              ("log", LogisticRegression(class_weight='balanced',
                                         random_state=101))]

model = Pipeline(steps=operations)


log_model_grid = GridSearchCV(model,
                              param_grid,
                              scoring=recall_Hispanic, #hispaniclerin recalln max yapacak ekilde hyper parameterlar bul
                              cv=10,
                              n_jobs=-1,
                              return_train_score=True)
```

```python
log_model_grid.fit(X_train,y_train) #modelimizi eittik
```

```python
log_model_grid.best_estimator_ #best estimatorlarna baktk
```

```python
pd.DataFrame(log_model_grid.cv_results_).loc[log_model_grid.best_index_, ["mean_test_score", "mean_train_score"]] #burada varyans azalm m ona bakyoruz ama classification zelinde sadece varyans ile overfite karar veremeyiz sadece recall sadece f1 e bakarak karar veremeyiz hem precision hem recalle bakmalyz gridsearch sonras alacam deerler iin tekrar bakmalyz 
```

```python
eval_metric(log_model_grid, X_train, y_train, X_test, y_test)
```

ama handikaplar var oversampling yazarken overfit yaparz o yzden azar azar artrma yapmalyz undersampling de ise datadan veri kayb yayrouz o yzden undersampling yaprken maks datann %20-25 ine kadar yapmalyz hala dzelmiyorsa under kullanmay devam etme nk veri kayb ok olacak dnyadan uzaklam olacaz over under yaptk dzeltemedik bitiriyoruz illa yapmak zorudnda deiliz ayrca KESNLKLE NCE TRANE AYIRIP SDCNTRAN TESTE UNDER OVER A YAPACAIZ NCE AYIRACAIZ TRAIN-TEST, SONRA TRAINI ALIP UNDER/OVER YAPACAIZ.hispaniclerin data iinde says az dierleri fazla. bu endenle yeteri kadar obsevation yok  feature bana den say yetersiz bunu artrmak iin oversampling/undersampling ilemleri

```python
from imblearn.over_sampling import SMOTE # aznlk olan classlar ounluk classa eitler veya yaknlatrr.
from imblearn.under_sampling import RandomUnderSampler # ounluk olan class aznlk olan classa eitler veya yaknlatrr
from imblearn.pipeline import Pipeline as imbpipeline#imblearnin pipeline da ayn oldupun as verdik
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore", 
                                                      sparse=False), cat), 
                                       remainder=MinMaxScaler())
```

```python
over = SMOTE()
X_train_over, y_train_over = over.fit_resample(X_train_ohe, y_train) #arka planda KNN kullanyor

# over_sampling(smote) ve under_sampling (RandomUnderSampler) sadece X_train ve y_train'e uygulanr. tm dataya uygulanp sonra
# data train ve test setine ayrlrsa data_leakege gereekleir ve scorlarnz ok iyi olur. Bu sebeple data nce train ve 
# test setine ayrlmal ve sonra da sadece train datasna uygulanmaldr.
```

```python
under = RandomUnderSampler()
X_train_under, y_train_under = under.fit_resample(X_train_ohe, y_train)
```

## Logistic Regression Over/ Under Sampling

imdi yukarda yaptmz tm dnmleri ve modele verme ilemlerini otomotize hale getirelim.

What does the smote pipeline do?

smote_pipeline.fit(X_train, y_train) 

--> column_trans.fit_transform(X_train) #(Onehooencoder and minmaxscaler for numeric features)

--> over.fit_resample(X_train_transform, y_train) 
                                     
--> under.fit_resample(X_train_transform_over, y_train_over)

--> log_model.fit(X_train_transform_over_under, y_train_over_under)


for predict, over and under sumpling algortims do nothing for X_test.

smote_pipeline.predict(X_test)

--> column_trans.transform(X_test) #(Onehotoencoder and minmaxscaler for numeric features) 

--> log_model.predict(X_test_transform)

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",
                                                      sparse=False), cat), 
                                       remainder=MinMaxScaler())
```
